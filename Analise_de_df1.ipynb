# Iniciar o uso de filtros para chegar a rela√ß√£o de Pa√≠ses, ano e produ√ß√£o de co2 para determinar os Pa√≠ses mais poluentes

```python
co2_df['country'].unique()
```

```python
len(co2_df['country'].unique())
```

```python
paises = co2_df['country'].unique().tolist()
```

```python
paises
```

```python
num_paises = len(paises)
num_paises
```

```python
co2_df[co2_df['country'] == paises[0]]
```

```python
co2_df[co2_df['country']==paises[0]]['year'].max()
```

```python
co2_df[co2_df['year']==co2_df[co2_df['country']==paises[0]]['year'].max()]
```

```python
dff=pd.DataFrame()
dff
```

```python
for i in range(num_paises):
  dftemp = co2_df.loc[(co2_df['country']==paises[i])&(co2_df['year']==co2_df[co2_df['country']==paises[i]]['year'].max())]
  dff = pd.concat([dff,dftemp], ignore_index= True)
  
```

```python
dff
```
